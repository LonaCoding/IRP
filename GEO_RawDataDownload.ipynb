```python
import os
import wget
import numpy as np
import pandas as pd
import re

dir='VPA/'

try: os.mkdir(dir+'rawData') 
except: pass

```

```python
# getting info about platforms to make sure we only downloading RNA expression data


import urllib.request, urllib.parse, urllib.error

# combining meta-data for each GSM
gsm_ds=pd.read_csv(dir+'Info/GSM-Meta_combined.csv')

gpls=gsm_ds['GPL'].unique()

gpl_descriptions=list()

for gpl in gpls:
    gpl=gpl.replace('\'','')
    gpl_info = urllib.request.urlopen('https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc='+gpl)
    for line in gpl_info:
        if b'<td nowrap>Title</td>' in line:
            array_info=next(gpl_info).strip().decode('utf-8')
            array_info=array_info.replace('<td style="text-align: justify">','')
            array_info=array_info.replace('</td>','')
            gpl_descriptions.append(array_info)
            
gplsDF=pd.DataFrame({'GPL':gpls,'Description':gpl_descriptions})     
gsm_ds2=pd.merge(gplsDF,gsm_ds,how='outer',on='GPL')
gsm_ds2.to_csv(dir+'Info/GSM-Meta_combined_annotated_withGPLs.csv')
```

```python
gsm_ds=pd.read_csv(dir+'Info/GSM-Meta_combined_annotated_withGPLs_adjusted.csv')
gsm_ds=gsm_ds[gsm_ds['Include']=='Yes']

gse_ds=pd.read_csv(dir+'Info/vpa_search_f1.csv')
gses=gsm_ds['GSE'].unique()
```

```python
for GSE in gses:
    os.mkdir(dir+'rawData/'+GSE)
    #os.chdir(dir+'rawData/'+GSE)
    folderBase=re.findall('^.+/(?=m)',gse_ds[gse_ds['GSE']==GSE]['DataLink'].tolist()[0])
    try:
        #wget.download(folderBase[0]+'/suppl/','files.txt')
        wget.download(folderBase[0]+'/suppl/',dir+'rawData/'+GSE+'/files.txt')
        with open(dir+'rawData/'+GSE+'/files.txt') as files:
            for line in files:
                file=re.findall('(?<=\s)\S*?$',line)[0]
                print('\nNow downloading '+GSE)
                #wget.download(folderBase[0]+'suppl/'+file)
                wget.download(folderBase[0]+'suppl/'+file,dir+'rawData/'+GSE)
        try:
            os.remove(dir+'rawData/'+GSE+'/files.txt')
            os.remove(dir+'rawData/'+GSE+'/filelist.txt')
        except: continue
    ##change back dir
    #os.chdir(dir+'rawData/'+GSE)
    except:
        continue
```
